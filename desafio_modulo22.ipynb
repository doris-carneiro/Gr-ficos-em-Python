```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv('C:/Users/Doris/Downloads/ecommerce_estatistica.csv')
print(df.head().to_string())

# Histograma - Par√¢metros
# para entender a distribui√ß√£o das principais m√©tricas
plt.figure(figsize=(10, 6))
plt.hist(df['Nota'], bins=100, color='green', alpha=0.8)
plt.title('Histograma - Distribui√ß√£o das Notas dos Produtos')
plt.xlabel('Notas')
plt.ylabel('Frequ√™ncia')
plt.show()

# Gr√°fico de Dispers√£o
# para entender a rela√ß√£o entre vari√°veis num√©ricas
plt.scatter(df['Pre√ßo'], df['Qtd_Vendidos_Cod'], color='#5883a8', alpha=0.6, s=30)    #cor hexadecimal online
plt.title('Disper√ß√£o entre Pre√ßo e Quantidade Vendida')
plt.xlabel('Pre√ßo (R$)')
plt.ylabel('Quantidade Vendida')
plt.show()

# Mapa de Calor
# para entender correla√ß√µes entre vari√°veis num√©ricas
corr = df[['Pre√ßo', 'N_Avalia√ß√µes']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correla√ß√£o Pre√ßo e N_Avalia√ß√µes')
plt.show()

# Gr√°fico de barras
# comparar categorias ou grupos de forma clara e visual
x = df['G√™nero'].value_counts().index
y = df['G√™nero'].value_counts().values
plt.figure(figsize=(10, 6))
plt.bar(x, y, color='skyblue')
plt.title('Qtde Vendida por G√™nero')
plt.xlabel('G√™nero')
plt.ylabel('Quantidade Vendida')
plt.xticks(rotation=30, ha='right')     # Rotaciona os nomes no eixo X
plt.show()

# Gr√°fico de pizza
# ideal para visualizar propor√ß√µes
plt.figure(figsize=(8, 6))
plt.pie(y,
    labels=x,
    autopct='%1.1f%%',       # mostra os percentuais
    startangle=90)            # come√ßa o gr√°fico no topo (90¬∞)
plt.title('Distribui√ß√£o de Vendas por G√™nero')
plt.show()


# Gr√°fico de Densidade
# Visualizar a distribui√ß√£o cont√≠nua de uma vari√°vel
# e entender onde h√° maior concentra√ß√£o de valores.
plt.figure(figsize=(10, 6))
sns.kdeplot(df['Pre√ßo'], fill=True, color='royalblue')
plt.title('Distribui√ß√£o de Densidade dos Pre√ßos dos Produtos')
plt.xlabel('Pre√ßo (R$)')
plt.ylabel('Densidade')
plt.show()

# Gr√°fico de Regress√£o
# analisar rela√ß√µes entre vari√°veis num√©ricas
plt.figure(figsize=(10, 6))
sns.regplot(x='Pre√ßo', y='Qtd_Vendidos_Cod', data=df, scatter_kws={'alpha':0.7, 'color':'royalblue'})
plt.title('Rela√ß√£o entre Pre√ßo e Quantidade Vendida')
plt.xlabel('Pre√ßo (R$)')
plt.ylabel('Quantidade Vendida')
plt.show()

```
